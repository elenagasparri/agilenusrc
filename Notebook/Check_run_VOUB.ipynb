```python
import os
import numpy
import matplotlib
```

```python
def list_files_recursive(path):
    """
    Function that receives as a parameter a directory path
    :return list_: File List and Its Absolute Paths
    """
    files = []
    # r = root, d = directories, f = files
    for r, d, f in os.walk(path):
          for direc in d:
                new_path=os.path.join(path,direc)
                for r,d,f in os.walk(new_path):
                    for file in f:
                        if '.log' in file:
                            files.append(os.path.join(r,file))
    return files
```

```python
def list_files_recursive(path,word):
    """
    Function that receives as a parameter a directory path and a string to serach for
    :return list_: File List and Its Absolute Paths
    """
    files = []
    # r = root, d = directories, f = files
    for r, d, f in os.walk(path):
          for direc in d:
                new_path=os.path.join(path,direc)
                for r,d,f in os.walk(new_path):
                    for file in f:
                        if word in file:
                            files.append(os.path.join(r,file))
    return files
```

```python
path = '/home/elenagasparri/VOU_Blazars/Results'
word = 'output1.csv'
```

```python
outfile=list_files_recursive(path,word)
```

```python
print(outfile)
```

```python
bad_words = ['BZQ','BZB','BZG','BZU','BZCAT']
good_run = []
bad_run=[]

for out in outfile:
    with open(out) as file:
        bad = False
        for line in file:
            if any(bad_word in line for bad_word in bad_words):
                bad=True
        if bad:
            bad_run.append(out)
        else:
            good_run.append(out)
```
